#### Setting the stage

```python
headers = {
    'User-Agent': 'Mozilla/5.0 (iPad; CPU OS 12_2 like Mac OS X) AppleWebKit/605.1.15 (KHTML, like Gecko) Mobile/15E148'
}
response = requests.get("https://download.bls.gov/pub/time.series/sm/sm.industry", headers = headers)
```

```python
#separating 
def separating(x, separator, confirm, start, end):
    x = x
    
    for sep in separator:
        x = x.replace(sep, "|")
    x = x.split("|")
    
    if (confirm == "Y"):
        del x[start: end]

    return(x)

#making dictionary keys and values(lists)
def function_name(x, dict_x, key_value, steps, len_set):
    dict_x[key_value] = []

    for i in range(0, len(x)-len_set):
        if i % 3 == 0:
            dict_x[key_value].append(x[i+steps])

    return(dict_x)
```

##### [Source](https://stackoverflow.com/questions/55848037/split-values-in-list-based-on-multiple-separators-python) for below solution

```python
x = separating(response.text, ["\r","\t","\n"], "Y", 0, 2)

emp_types = {}
step = 1
et_times = ["industry_code", "industry_name"]

for i in range(0, len(et_times)):
    emp_types = function_name(x, emp_types, et_times[i], step, 2)
    step += 1

emp_types = pd.DataFrame(emp_types)
emp_types
```

```python
response = requests.get("https://download.bls.gov/pub/time.series/la/la.area", headers = headers)
l = response.text
```

```python
separator = ["\r","\t","\n"]
for sep in separator:
    l = l.replace(sep, "")

list_states = re.findall("[AST]+[0-6][0-9]0{11}+[A-Za-z|\sA-Za-z]+[\d0T]", l)
```

```python
area_codes = {}
area_codes["area_codes"] = []
area_codes["states"] = []
area_codes["state_code"] = []

for element in list_states:
    var = str(element)
    area_codes["area_codes"].append(var[:16])
    area_codes["states"].append(var[16:-1])
    area_codes["state_code"].append(var[3:5])

area_codes = pd.DataFrame(area_codes)
area_codes
```

```python
area_codes.to_excel("bls_survey_codes.xlsx", sheet_name= "State Codes")
```

```python
def insert_into_wrbk(df, path, sheet_name):
    with pd.ExcelWriter(path, engine = "openpyxl") as writer:
        writer.book = load_workbook(path)
        df.to_excel(path, sheet_name, cols = list(df.columns))
```

```python
list_df = [[area_codes, "State Codes"], [emp_types, "Industry Codes"]]

for df in list_df:

    insert_into_wrbk(df[0], "bls_survey_codes.xlsx", df[1])
```
