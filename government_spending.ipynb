```python
stuff = ba.collection([{"survey" : "NIPA", "table" : "T80104", "freq" : "A"}])
stuff
```

```python
description = stuff["LineDescription"].unique()
concat = pd.DataFrame()        

for item in description:
    df = stuff[stuff["LineDescription"] == item]
    df = df.pivot_table(index = "dates", columns = "LineDescription", values = "DataValue", aggfunc = 'first')
    concat = pd.concat([concat, df], axis = 1)

concat.reset_index(inplace = True)
        
```
